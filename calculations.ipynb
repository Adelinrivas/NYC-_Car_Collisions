# Calculations

### By Borough

Get the number of accidents/collisions by borough

```python
boroughs_df = pd.DataFrame()
boroughs_df["BOROUGH"] = clean_df["BOROUGH"]
```

```python
boroughs_df.head(100)
```

```python
boroughs_df = boroughs_df.drop_duplicates()
```

```python
boroughs_df = boroughs_df.reset_index(drop=True)
```

```python
# Count how many accidents occur in these buroughs
results = []
for i in range(boroughs_df.shape[0]):
    counter = 0
    def countingAccidentsByBorough(row, borough: str):
        global counter
        if row["BOROUGH"] == borough:
            counter += 1
    clean_df.apply(countingAccidentsByBorough, axis=1, args=(boroughs_df.iloc[i]["BOROUGH"], ))
    results.append(counter)
    print("borough: ", boroughs_df.iloc[i]["BOROUGH"], "; total accidents: ", counter)
    
tempDf = pd.DataFrame(results, columns=["NUMBER_OF_ACCIDENTS_BOROUGH"])
boroughs_df["NUMBER_OF_ACCIDENTS_BOROUGH"] = tempDf["NUMBER_OF_ACCIDENTS_BOROUGH"]
```

```python
boroughs_df.head(5)
```

## About the Vehicle(s)

```python
tempDf1 = pd.DataFrame()
tempDf2 = pd.DataFrame()

tempDf1["VEHICLES"] = clean_df["VEHICLE TYPE CODE 1"]
tempDf2["VEHICLES"] = clean_df["VEHICLE TYPE CODE 2"]
```

```python
vehicles_df = pd.concat([tempDf1, tempDf2], ignore_index=True)
```

```python
vehicles_df = vehicles_df.dropna()
```

```python
vehicles_df = vehicles_df.drop_duplicates()
```

```python
vehicles_df = vehicles_df.reset_index(drop=True)
```

```python
vehicles_df.shape
```

## Takes a long time!

Count the number of collisions by vehicle type

```python
# Count how many accidents occur by vehicle type
# Because there are two columns for vehicles, you'll need to take both columns in
startTime = time.time()


results = []
for i in range(vehicles_df.shape[0]):
    counter = 0
    def countingAccidentsByVehicle(row, vehicle: str):
        global counter
        if row["VEHICLE TYPE CODE 1"] == vehicle and row["VEHICLE TYPE CODE 2"] == vehicle:
            #print("\nvehicle: ", vehicles_df.iloc[i]["VEHICLES"], "\nvehicle 1: ", row["VEHICLE TYPE CODE 1"], "\nvehicle 2: ", row["VEHICLE TYPE CODE 2"], "\ncounting up 2", "\n========")
            counter += 2
        elif row["VEHICLE TYPE CODE 1"] == vehicle or row["VEHICLE TYPE CODE 2"] == vehicle:
            #print("\nvehicle: ", vehicles_df.iloc[i]["VEHICLES"], "\nvehicle 1: ", row["VEHICLE TYPE CODE 1"], "\nvehicle 2: ", row["VEHICLE TYPE CODE 2"], "\ncounting up 1", "\n========")
            counter += 1
        
    clean_df.apply(countingAccidentsByVehicle, axis=1, args=(vehicles_df.iloc[i]["VEHICLES"], ))
    results.append(counter)
    print("index: ", i, "; vehicle: ", vehicles_df.iloc[i]["VEHICLES"], "; total accidents: ", counter)
    
tempDf = pd.DataFrame(results, columns=["NUMBER_OF_ACCIDENTS_VEHICLE"])
vehicles_df["NUMBER_OF_ACCIDENTS_VEHICLE"] = tempDf["NUMBER_OF_ACCIDENTS_VEHICLE"]


executionTime = (time.time() - startTime)
print('Execution time in seconds: ' + str(executionTime))
```

```python
vehicles_df.to_csv("vehicles_calculations.csv")
```

```python
reasons_df = pd.concat([tempDf1, tempDf2], ignore_index=True)
```

```python
reasons_df = reasons_df.dropna()
```

```python
reasons_df = reasons_df.drop_duplicates()
```

```python
reasons_df = reasons_df.reset_index(drop=True)
```

```python
reasons_df.head()
```

```python
# Count how many accidents occur by reason type
# Because there are two columns for reasons, you'll need to take both columns in
startTime = time.time()


results = []
for i in range(reasons_df.shape[0]):
    counter = 0
    def countingReasons(row, reason: str):
        global counter
        if row["CONTRIBUTING FACTOR VEHICLE 1"] == reason and row["CONTRIBUTING FACTOR VEHICLE 2"] == reason:
            #print("\nreason: ", reasons_df.iloc[i]["REASONS"], "\nreason 1: ", row["CONTRIBUTING FACTOR VEHICLE 1"], "\nreason 2: ", row["CONTRIBUTING FACTOR VEHICLE 2"], "\ncounting up 2", "\n========")
            counter += 2
        elif row["CONTRIBUTING FACTOR VEHICLE 1"] == reason or row["VEHICLE TYPE CODE 2"] == reason:
            #print("\nreason: ", reasons_df.iloc[i]["REASONS"], "\nreason 1: ", row["CONTRIBUTING FACTOR VEHICLE 1"], "\nreason 2: ", row["CONTRIBUTING FACTOR VEHICLE 2"], "\ncounting up 1", "\n========")
            counter += 1
        
    clean_df.apply(countingReasons, axis=1, args=(reasons_df.iloc[i]["REASONS"], ))
    results.append(counter)
    print("reason: ", reasons_df.iloc[i]["REASONS"], "; total: ", counter)
    
tempDf = pd.DataFrame(results, columns=["TOTAL"])
reasons_df["TOTAL"] = tempDf["TOTAL"]


executionTime = (time.time() - startTime)
print('Execution time in seconds: ' + str(executionTime))
```

```python
reasons_df.to_csv("reasons_for_collisions.csv") 
```

## By Year

How many collisions by year?

```python
split_clean_df = clean_df.copy()
```
